```python
import pandas as pd

results_dir = '~/Programming/shipping_sprint/gpo_transitions/results/'
df = pd.read_excel(results_dir+'uk_added_demand/2050_sc4.5_df1_fpdf1_ukes.xlsx',sheet_name='Demand')

# Filter rows where 'Port' ends with '_United Kingdom'
filtered_df = df[df['Port'].str.endswith('_United Kingdom')]

filtered_df
```

```python
# Total country demand
# Calculate the sum of demand multiplied by LCOA
demand_sum = filtered_df['Demand (MMTPA)'].sum()
weighted_lcoa_sum = (filtered_df['Demand (MMTPA)'] * filtered_df['Delivered Cost']).sum()

# Calculate the overall LCOA
overall_lcoa = weighted_lcoa_sum / demand_sum

# Print the overall demand and LCOA
print("Overall demand:", demand_sum)
print("Overall LCOA:", overall_lcoa)
```
